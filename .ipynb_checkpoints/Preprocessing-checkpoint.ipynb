### Investigating numerical attributes

First, we take a look at the features values distributions in order to have a first insight at their behavior. We are going to split these attributes in a couple of groups, according to their perceived affinity. 

```python
#Define functions for grouped distribution plots
def hist_subplots(data, cols, nrows, ncols):
    fig, axes = plt.subplots(nrows, ncols, figsize = (14,6))
    plt.subplots_adjust(wspace = 0.3, hspace = 0.40)
    for i,col in enumerate(cols):
        nr = i // ncols
        nc = i - nr * ncols
        sns.distplot(data[col].values, ax = axes[nr,nc]).set_title(col)
    plt.show()
    
def count_subplots(data, cols, nrows, ncols):
    fig, axes = plt.subplots(nrows, ncols, figsize = (14,6))
    plt.subplots_adjust(wspace = 0.3, hspace = 0.40)
    for i,col in enumerate(cols):
        nr = i // ncols
        nc = i - nr * ncols
        sns.countplot(data[col].values, ax = axes[nr,nc]).set_title(col)
    plt.show()
```

### Investigate categorical attributes

The categorical attributes are explored by group, in a similar approach to the one adopted for the numeric attributes. In order to obtain a estimation of how well each attribute relates directly to the target value, we perform a simple regression on the encoded categorical features. The relationship is then measured by the R score.

```python
#Define functions for categorical attributes exploration
def cat_regression(data, cols, target):
    score = []
    for col in cols:
        dummy = pd.get_dummies(data[col])
        model = Ridge(normalize = True)
        model.fit(dummy.values, data[target].values)
        score.append(model.score(dummy.values, data[target].values))
    return pd.Series(data = score, index = cols)
```

# Outliers Detection

Before we advance to our regression model, it is important to investigate ou dataset in order to identify outliers. These instances, however few they might be, can have a significant negative effect on the quality of our predictions.

```python
#Find outliers
outliers = []
for line in tqdm(range(len(train_data))):
    count = 0
    for col in num_atr:
        try:
            val = train_data[col].values
            if abs(val.mean() - val[line]) > 2.0 * val.std():
                count += 1
        except:
            pass
    outliers.append(count)

outliers = np.array(outliers)
```

```python
#Print outliers profile
pd.DataFrame(data = np.unique(outliers, return_counts = True), index = ['AmountOfOutliers', 'NoInstancies'])
```

The definition of what consists in a outlier instancies in a complex dataset is actually aritrary. Here, all instancies with more than 4 outliyng features are removed from the train dataset.

```python
#Remove outliers
train_data = train_data.iloc[outliers <= 4,:]
```

# Regression model

For the regression model, we are going to apply a Multilayer Perceptron network using the Keras library.

The first step is then to define the neural network as a function.

```python
#Define neural network as function
def mlp_regression(n_in, n_out, n_nodes, n_layers, act, loss_f, opt, drop):
  winit = 'glorot_uniform'  
    
  model = Sequential()
  model.add(Dense(units = n_nodes, activation = act, kernel_initializer = winit, input_dim = n_in))
  model.add(Dropout(drop))
  for i in range(n_layers):
    model.add(Dense(units = n_nodes, activation = act, kernel_initializer = winit))
    model.add(Dropout(drop))
  model.add(Dense(units = n_out, activation = 'linear', kernel_initializer = winit))
  
  model.compile(optimizer = opt, loss = loss_f, 
                metrics = ['mean_squared_logarithmic_error'])
  return model
```

```python
#Split train dataset into target and features
target = train_data['SalePrice'].values
atr = train_data.drop(columns = 'SalePrice').values
```

```python
#Test model
first_model = KerasRegressor(build_fn = mlp_regression,
                             n_in = train_data.shape[1]-1,
                             n_out = 1,
                             n_nodes = 50,
                             n_layers = 1,
                             act = 'relu',
                             loss_f = 'mean_squared_logarithmic_error',
                             opt = 'adam',
                             drop = 0.2,
                             batch_size = 32,
                             epochs = 500)

#hist = first_model.fit(atr, target, epochs = 200, batch_size = 32)
score = cross_val_score(estimator = first_model,
                        X = atr,
                        y = target,
                        scoring = 'neg_mean_squared_log_error',
                        cv = 10,
                        n_jobs = 4,
                        verbose = 0)
```

```python
#Print first test score
print(-score.mean())
```

## Tuning hyperparameters

In order to guarantee optimal performance, we first have to tune the regression model hyperparameters. To do that, we must write our model in the following form:

```python
model = KerasRegressor(build_fn = mlp_regression)
```

The next step is to set the hyperparameters grid. 

```python
param = {'n_in': [atr.shape[1]],
         'n_out': [1],
         'n_layers': [1, 2, 3, 4],
         'n_nodes': [30, 50, 80, 100, 150, 200],
         'opt': ['adam'],
         'act': ['relu'],
         'loss_f': ['mean_squared_logarithmic_error'],
         'drop': [0.2],
         'batch_size': [32],
         'epochs': [500]}
```

```python
grid = GridSearchCV(estimator = model,
                    param_grid = param,
                    scoring = 'neg_mean_squared_log_error',
                    cv = 10,
                    n_jobs = 2,
                    verbose = 2)

best_params = []
for i in tqdm(np.unique(groups)):
  grid.fit(atr[groups == i], target[groups == i])
  best_params.append(grid.best_params_)
```
